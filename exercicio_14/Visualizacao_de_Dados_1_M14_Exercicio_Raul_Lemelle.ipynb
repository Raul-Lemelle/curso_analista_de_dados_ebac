## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
import matplotlib.pyplot as plt

''' Informações do dataset "diamonds":
- price in US dollars (\$326 — \$18,823)
- carat weight of the diamond (0.2–5.01)
- clarity a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
'''

# Colunas de interrese para análise prévia:
diamonds = data[['carat', 'price','clarity']].groupby('price')
diamonds.head()
```

```
# Gráfico:
grafico = sns.relplot(data=data, x='carat', y='price', hue='clarity', style='clarity')
grafico.set(title='Diamantes: Quilates vs Preço - Baseado na Clareza', xlabel='Quilates', ylabel='Preço (USD)')
grafico._legend.set_title("Clareza")
```

**Insight do gráfico 1**: Com base na análise do gráfico acima, observa-se que o preço de diamantes com menor peso em quilates tende a ser mais elevado quando sua clareza é superior. Por outro lado, diamantes com menor clareza geralmente demandam um peso maior em quilates para alcançar valores similares.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data
```

```
# gráfico do exercício 2
flights = data.query('month == "Dec"').reset_index()
flights.head()
```

```
grafico = sns.barplot(data=flights, x='year', y='passengers', palette='dark')
grafico.set(title='Passageiros em dezembro por ano', xlabel='Ano', ylabel='Passageiros')
sns.despine() # remoção das bordas.
```

**Insight do gráfico 2**: Com base na análise do gráfico, observa-se um contínuo aumento no número de passageiros durante o mês de dezembro ao longo dos anos, indicando que, aproximadamente a cada 5 anos, ocorre uma duplicação no volume de passageiros.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3
import matplotlib.pyplot as plt

flights = data.query('1949 <= year < 1960').reset_index()
flights.head()
```

```
grafico = sns.barplot(data=flights, x='year', y='passengers', hue='month', palette='bright')
grafico.set(title='Passageiros por mês entre 1949 e 1959', xlabel='Ano', ylabel='Passageiros')
sns.despine() # remoção das bordas.
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') # legenda fora do gráfico
```

**Insight do gráfico 3**: Com base no gráfico, é possível inferir que, ao longo dos anos, os meses de julho e agosto destacaram-se como os mais movimentados no setor da aviação, sugerindo que este período coincide com as férias de meio de ano.
